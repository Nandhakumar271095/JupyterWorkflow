```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
df = pd.read_csv("Finance_data.csv")
df.head()
```

```python
df.drop(['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold'],axis=1,inplace=True)
```

```python
sns.countplot(df['gender'], linewidth=2, palette='Set2', edgecolor='white')
plt.show()
```

```python
sns.countplot(df['age'], linewidth=3, palette='Set3', edgecolor='black')
plt.show
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['gender'], hue=df['Investment_Avenues'], palette='summer', linewidth=3, edgecolor='black')
plt.title('INVESTMENT AVENUES')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'], hue=df['Stock_Marktet'],palette='hot',linewidth=3,edgecolor='black')
plt.title('STOCK MARKET')
plt.show()
```

```python
sns.countplot(x=df['Factor'],palette='coolwarm',linewidth=2,edgecolor='black')
```

```python
sns.countplot(x=df['gender'], hue=df['Factor'],palette='Oranges',linewidth=2, edgecolor='black')
plt.title('Gender')
plt.show()
```

```python
plt.figure(figsize=(17,6))
sns.pointplot(x='Purpose', y='age', data=df, linestyles='--', capsize=.3, color='red')
plt.show()
```

```python
dummies = pd.get_dummies(df['gender'])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
Investment_Avenues_dummies = pd.get_dummies(df['Investment_Avenues'])
Stock_Marktet_dummies = pd.get_dummies(df['Stock_Marktet'])
Investment_Avenues_dummies
Stock_Marktet_dummies
```

```python
merged_final = pd.concat([df, Investment_Avenues_dummies, Stock_Marktet_dummies, dummies], axis='columns')
merged_final
```

```python
final = merged_final.drop(['gender', 'Investment_Avenues', 'Stock_Marktet', 'Factor', 'Objective', 'Purpose', 'Duration'
                          ,'Invest_Monitor', 'Expect', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD','Source'], axis='columns')

final
```

```python
final_1 = final.drop(['Avenue', 'What are your savings objectives?', 'Reason_Equity'], axis='columns')
final_1
```

```python
X = final_1.drop(['age'], axis='columns')
X
```

```python
y = final_1['age']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```
